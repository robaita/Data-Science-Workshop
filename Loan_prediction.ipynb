# Loan Prediction Model
**Problem Statement:- A finance company is receiving lots of loan applications and the employees were manualy accepting and rejecting the applications based on certain criteria. The company wants to automate the process and auto reject and accept the applications**.

**Solution:- To solve the above problem we can train a machine learning model using the past data of the organisation and based on that the model can predict whether the customer is eligible for home loan or not.**

# **Step 1:- Importing necessary libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 4:- Data cleaning and imputing missing values

```python
#filling the null values by calculating mode of the columns
df.Gender = df.Gender.fillna(df.Gender.mode()[0])
df.Married = df.Married.fillna(df.Married.mode()[0])
df.Dependents = df.Dependents.fillna(df.Dependents.mode()[0])
df.Self_Employed = df.Self_Employed.fillna(df.Self_Employed.mode()[0])
df.LoanAmount = df.LoanAmount.fillna(df.LoanAmount.mode()[0])
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.mode()[0])
df.Credit_History = df.Credit_History.fillna(df.Credit_History.mode()[0])


```

```python
df.isnull().sum()
```

```python
#checking the unique values in all the columns
df.nunique()
```

```python
#removing the Loan_ID column as it is of no significance to us
data = df.drop(columns= ['Loan_ID'], inplace= False)
```

```python
display(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

# Labelling the categorical values with 0 and 1

```python
cat_cols = ['Dependents','Gender','Married','Education','Self_Employed','Credit_History','Property_Area']
dummies = pd.get_dummies(data[cat_cols])
dummies
```

```python
#adding labelled columns to original data frame
data= pd.concat([data, dummies], axis = 1)
```

```python
#removing categorical columns from original data frame
data = data.drop(cat_cols, axis = 1)
```

```python
display(data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#encoding the Loan_status column with 0 and 1, 0 means NO, 1 means yes
le = LabelEncoder()
y = le.fit_transform(data['Loan_Status'])
y
```

# Step 5: Training the machine learning model using logistic Regression

```python
#creating training and testing dataset
x = data.drop('Loan_Status', axis =1)#input variables
y = data['Loan_Status']#target variable
print(x.shape)
print(y.shape)
```

```python
#splitting the dataset into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
#applying the Logistic Regression on the training data
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#making predctions
pred=model.predict(x_test)
pred
```

```python
#Cheking the accuracy  of the model
print("accuracy score : ",accuracy_score(y_test,pred))
```
