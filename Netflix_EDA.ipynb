# Solving first question:-
How much of the Netflix content are TV shows or Movies?

**let's create a pie chart to see the dividation of TV shows and movies in the whole content**

```python
plt.figure(figsize=(12,6)) #defining the size of plot
labels = ['Tv show', 'Movie']
plt.pie( df['type'].value_counts().sort_values(), labels=labels,explode=[0.1,0.1],autopct='%1.2f%%', colors=['blue','red'], startangle=90)
plt.axis('equal')
plt.title("Types of Netflix content")
plt.show()
```

from the above pie plot we can easily identify that most of the content(about 70%) on Netflix are movies.

# Solving second question:-
In which country is Netflix the most popular?
We can find it by counting the number of viewers in every country.

```python
#counting number of viewers
country_df= df['country'].value_counts().reset_index()
country_df
```

we can see there are 748 total places where netflix is viewed but we can also see there are many places having very less number of viewers. so we will only take those places in account which have viewership of more than 1%.

```python
country_df= country_df[country_df['country']/ country_df['country'].sum()>0.01]
country_df
```

```python
#lets plot the above data
plt.figure(figsize=(14,7))
labels = country_df['index']
plt.pie(country_df['country'], labels=labels)
plt.axis('equal')
plt.title("Netfilx viewers across the globe")
plt.show()
```

From this pie chart we can conclude that USA, India and UK are the top three countries which have the largest audience of Netflix.
